```python



import matplotlib.pyplot as plt

# Simple data
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

# Create a line plot
plt.plot(x, y, color='red', marker='s', linestyle='--', linewidth=2, markersize=8)

# Add labels and title
plt.xlabel("X-axis (Numbers)")
plt.ylabel("Y-axis (values)")
plt.title("Simple Line Plot")

# Show the plot
plt.show()
```

```python
x = ['Apple', 'Banana', 'Grapes', 'Mango']
y = [30, 50, 20, 60]
plt.bar(x, y, color='skyblue')
plt.xlabel("Fruits")
plt.ylabel("Quantity")
plt.title("Fruit Quantity Bar chart")
plt.show()
```

```python
labels = ['Python', 'Java', 'C++', 'Javascript']
sizes = [40, 25, 20, 15]
colors = ['gold', 'lightcoral', 'lightblue', 'yellowgreen']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90)
plt.title("Programming Language Popularity")
plt.show()
```

```python
x = [5, 7, 8, 9, 10, 12]
y = [12, 14, 17, 20, 23, 27]

plt.scatter(x, y, color='purple')
plt.xlabel("Age")
plt.ylabel("Score")
plt.title("Age vs Score Scatter Plot")
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y1 = [10, 15, 20, 25, 30]
y2 = [8, 13, 18, 23, 28]

plt.plot(x, y1, label="Data 1", color='blue', marker='o')
plt.plot(x, y2, label="Data 2", color='red', marker='x')
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Multiple Line Plot")
plt.legend() # show labels
plt.show()
```

```python
import numpy as np

data = np.random.randn(1000)

plt.hist(data, bins=20, color='orange', edgecolor='black')
plt.title("Histogram Example")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```
